```python
# --- DEFINISI LOSS DAN MODEL VGG19 ---

# Konstanta normalisasi untuk model VGG19
cnn_normalization_mean = torch.tensor([0.485, 0.456, 0.406]).to(device)
cnn_normalization_std = torch.tensor([0.229, 0.224, 0.225]).to(device)

class Normalization(nn.Module):
    """Layer normalisasi untuk input gambar agar sesuai dengan training VGG."""
    def __init__(self, mean, std):
        super(Normalization, self).__init__()
        # Membuat tensor 4D [1, C, 1, 1] agar bisa digunakan dalam broadcasting
        self.mean = mean.clone().detach().view(-1, 1, 1)
        self.std = std.clone().detach().view(-1, 1, 1)

    def forward(self, img):
        return (img - self.mean) / self.std

def gram_matrix(input):
    """Menghitung Gram Matrix untuk Style Loss (menangkap tekstur dan gaya)."""
    # input memiliki dimensi (batch_size, channels, height, width)
    a, b, c, d = input.size()
    features = input.view(a * b, c * d)
    G = torch.mm(features, features.t())
    # Normalisasi Gram Matrix
    return G.div(a * b * c * d)

# Loss Content
class ContentLoss(nn.Module):
    def __init__(self, target, weight):
        super(ContentLoss, self).__init__()
        self.target = target.detach()
        self.weight = weight

    def forward(self, input):
        self.loss = self.weight * nn.functional.mse_loss(input, self.target)
        return input

# Loss Style
class StyleLoss(nn.Module):
    def __init__(self, target_feature, weight):
        super(StyleLoss, self).__init__()
        self.target = gram_matrix(target_feature).detach()
        self.weight = weight

    def forward(self, input):
        G = gram_matrix(input)
        self.loss = self.weight * nn.functional.mse_loss(G, self.target)
        return input

# Model VGG19 (Menggunakan pretrained=True untuk kompatibilitas PyTorch lama/baru)
cnn = models.vgg19(pretrained=True).features.to(device).eval()

# Lapisan VGG yang digunakan untuk menghitung Loss
content_layers = ['conv_4']
style_layers = ['conv_1', 'conv_2', 'conv_3', 'conv_4', 'conv_5']


def get_model_and_losses(cnn, normalization_mean, normalization_std, style_img, content_img):
    """Membuat model Sequential dengan Normalisasi dan Loss Layers."""
    # Bobot Gaya (1e6) dan Konten (1e3) adalah rasio umum yang bisa disesuaikan
    style_weights = 1e6
    content_weights = 1e3

    cnn = copy.deepcopy(cnn)
    normalization = Normalization(normalization_mean, normalization_std).to(device)
    content_losses = []
    style_losses = []

    model = nn.Sequential(normalization)
    i = 0
    for layer in cnn.children():
        if isinstance(layer, nn.Conv2d):
            i += 1
            name = 'conv_{}'.format(i)
        elif isinstance(layer, nn.ReLU):
            name = 'relu_{}'.format(i)
            layer = nn.ReLU(inplace=False)
        elif isinstance(layer, nn.MaxPool2d):
            name = 'pool_{}'.format(i)
        elif isinstance(layer, nn.BatchNorm2d):
            name = 'bn_{}'.format(i)
        else:
            continue

        model.add_module(name, layer)

        if name in content_layers:
            target = model(content_img).detach()
            content_loss = ContentLoss(target, weight=content_weights)
            model.add_module("content_loss_{}".format(i), content_loss)
            content_losses.append(content_loss)

        if name in style_layers:
            target_feature = model(style_img).detach()
            style_loss = StyleLoss(target_feature, weight=style_weights)
            model.add_module("style_loss_{}".format(i), style_loss)
            style_losses.append(style_loss)

    # Potong layer setelah loss layer terakhir
    for i in range(len(model) - 1, -1, -1):
        if isinstance(model[i], ContentLoss) or isinstance(model[i], StyleLoss):
            break
    model = model[:(i + 1)]

    return model, style_losses, content_losses

# --- 3. PROSES OPTIMASI NST ---

# Gambar input awal (menggunakan content image sebagai starter)
input_img = content_img.clone()

def get_input_optimizer(input_img):
    # L-BFGS adalah optimizer yang paling efektif untuk masalah NST ini
    optimizer = optim.LBFGS([input_img.requires_grad_()])
    return optimizer

def run_style_transfer(cnn, normalization_mean, normalization_std,
                       content_img, style_img, input_img, num_steps=300):
    """Fungsi utama untuk menjalankan proses NST."""
    print("\nðŸ”¬ Membangun model style transfer...")
    model, style_losses, content_losses = get_model_and_losses(
        cnn, normalization_mean, normalization_std, style_img, content_img)

    optimizer = get_input_optimizer(input_img)

    print("ðŸš€ Mulai Optimasi. Proses ini membutuhkan waktu...")
    run = [0]
    while run[0] <= num_steps:

        def closure():
            # Mengoreksi nilai piksel yang keluar dari batas [0, 1]
            input_img.data.clamp_(0, 1)

            optimizer.zero_grad()
            model(input_img)

            style_score = 0
            content_score = 0

            for sl in style_losses:
                style_score += sl.loss

            for cl in content_losses:
                content_score += cl.loss

            # Total Loss
            loss = style_score + content_score
            loss.backward()

            run[0] += 1
            if run[0] % 50 == 0:
                print(f"Langkah {run[0]}/{num_steps}: Style Loss: {style_score.item():.4f} Content Loss: {content_score.item():.4f}")

            return style_score + content_score

        optimizer.step(closure)

    # Klem hasil akhir
    input_img.data.clamp_(0, 1)

    return input_img

# Jalankan proses transfer gaya
output = run_style_transfer(cnn, cnn_normalization_mean, cnn_normalization_std,
                            content_img, style_img, input_img, num_steps=300)

# --- 4. SIMPAN HASIL ---

# Simpan gambar hasil
save_image(output, OUTPUT_IMAGE)

print(f"\nðŸŽ‰ Proses Selesai!")
print(f"Hasil penggabungan gaya disimpan di file: {OUTPUT_IMAGE}")

# Tampilkan gambar hasil di Colab (opsional)
try:
    plt.figure()
    plt.imshow(transforms.ToPILImage()(output.squeeze(0).cpu()))
    plt.title("Hasil Penggabungan Gaya")
    plt.axis('off')
    plt.show()
except Exception as e:
    print(f"Gagal menampilkan gambar: {e}")
```
